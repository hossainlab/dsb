# What is List
In short, a list is a **collection** of arbitrary objects, somewhat akin to an array in many other programming languages but more flexible. Lists are defined in Python by enclosing a comma-separated sequence of objects in square brackets ([])

## Characteristics of List
- Lists are ordered.
- Lists can contain any arbitrary objects.
- List elements can be accessed by index.
- Lists can be nested to arbitrary depth.
- Lists are mutable.
- Lists are dynamic.

````python
# Create a list: string 
fruits = ["apple", "orange", "banana"]
````

### Explanation of the Code

- **`fruits = ["apple", "orange", "banana"]`**:
  - This line creates a Python **list** named `fruits`.
  - The list contains **three string elements**: `"apple"`, `"orange"`, and `"banana"`.
  - **Lists** in Python are used to store multiple items in a single variable and are defined by **square brackets `[]`**.
  
### Key Concepts:
1. **List**:
   - A list is a **data structure** in Python that is used to store **multiple items**.
   - Lists are **ordered**, meaning the items have a defined order.
   - Lists are also **mutable**, which means you can modify their content (add, remove, or change elements).
   
2. **String Elements**:
   - Each item in the list is a **string** (`"apple"`, `"orange"`, `"banana"`).
   - Strings are **text data** enclosed in quotes.

3. **Accessing Elements**:
   - You can access elements in the list by their **index**. Indexing starts at `0` in Python.
   - Example:
     ```python
     print(fruits[0])  # Outputs: apple
     print(fruits[1])  # Outputs: orange
     print(fruits[2])  # Outputs: banana
     ```

4. **List Properties**:
   - **Ordered**: The position of each element matters. The first item is `"apple"`, followed by `"orange"`, and then `"banana"`.
   - **Mutable**: You can modify the list, for example:
     ```python
     fruits[0] = "grape"
     print(fruits)  # Outputs: ['grape', 'orange', 'banana']
     ```

**Modify a list (does not return the list):**

````python
# append element to end
simpsons.append('lisa') #append
simpsons
````

````python
# append multiple elements to end
simpsons.extend(['itchy', 'scratchy'])  #extend
simpsons
````

````python
# insert element at index 0 (shifts everything right)
simpsons.insert(0, 'maggie')   #insert 
simpsons
````

````python
# search for first instance and remove it
simpsons.remove('bart')    #remove
simpsons
````

````python
# remove element 0 and return it
simpsons.pop(0)   #pop
````

````python
# remove element 0 (does not return it)
del simpsons[0]
simpsons
````

````python
# replace element 0
simpsons[0] = 'krusty'
simpsons
````

````python
# concatenate lists (slower than 'extend' method)
neighbors = simpsons + ['ned', 'rod', 'todd']
neighbors
````

**Find elements in a list:**

````python
# counts the number of instances
simpsons.count('lisa')
````

````python
# returns index of first instance
simpsons.index('itchy')
````

**List slicing:**

````python
weekdays = ['mon', 'tues', 'wed', 'thurs', 'fri']
````

````python
# element 0
weekdays[0]
````

````python
# elements 0 (inclusive) to 3 (exclusive)
weekdays[0:3]
````

````python
# starting point is implied to be 0
weekdays[:3]
````

````python
# elements 3 (inclusive) through the end
weekdays[3:]
````

````python
# last element
weekdays[-1]
````

````python
# every 2nd element (step by 2)
weekdays[::2]
````

````python
# backwards (step by -1)
weekdays[::-1]
````

````python
# alternative method for returning the list backwards
list(reversed(weekdays))
````

**Sort a list in place (modifies but does not return the list):**

````python
simpsons.sort()
simpsons
````

````python
# sort in reverse
simpsons.sort(reverse=True)
simpsons
````

````python
# sort by a key
simpsons.sort(key=len)
simpsons
````

**Return a sorted list (does not modify the original list):**

````python
sorted(simpsons)
````

````python
sorted(simpsons, reverse=True)
````

````python
sorted(simpsons, key=len)
````

**Insert into an already sorted list, and keep it sorted:**

````python
num = [10, 20, 40, 50]
from bisect import insort
insort(num, 30)
num
````

**Object references and copies:**

````python
# create a second reference to the same list
same_num = num
````

````python
# modifies both 'num' and 'same_num'
same_num[0] = 0
print(num)
print(same_num)
````

````python
# copy a list (two ways)
new_num = num[:]
new_num = list(num)
````

**Examine objects:**

````python
num is same_num    # checks whether they are the same object
````

````python
num is new_num
````

````python
num == same_num    # checks whether they have the same contents
````

````python
num == new_num
````
